**Bar Chart for Gender Distribution**

```python
df['gender_label'] = df['gender'].map({0: 'Male', 1: 'Female'})

plt.figure(figsize=(7, 5))
sns.countplot(x='gender_label', data=df, palette='viridis') # Using 'gender_label' for plotting
plt.title('Distribution of Gender in Population', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Histogram for Age Distribution**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=range(0, df['age'].max() + 5, 5), kde=True, color='skyblue') # Bins of 5 years
plt.title('Distribution of Age in Population', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Bar Chart for each unique Age Distribution**

```python
top_ages = df['age'].value_counts().sort_index()

plt.figure(figsize=(15, 7))
sns.barplot(x=top_ages.index, y=top_ages.values, palette='magma')
plt.title('Count of Individuals per Age (Full Range)', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, fontsize=10) # Rotate labels if too many ages
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

**Bar Chart: Gender distribution by Age**

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='age', hue='gender_label', palette='coolwarm', order=df['age'].value_counts().sort_index().index)
plt.title('Gender Distribution Across Age Groups', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90, fontsize=9) # Rotate age labels for readability
plt.yticks(fontsize=10)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
